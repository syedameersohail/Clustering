```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cd=pd.read_csv('happyscore_income.csv',index_col=None)
cd=cd.sort_values(by=['GDP'],ascending=False)
cd.head()
```

```python
gdp=cd['GDP']
happysc=cd['happyScore']
```

```python
m1=gdp.mean()
maxg=gdp.max()
ming=gdp.min()
gdp_norm=(gdp-m1).div(maxg-ming)
gdp_norm
```

```python
m2=happysc.mean()
maxh=happysc.max()
minh=happysc.min()
happy_norm=(happysc-m2).div(maxh-minh)
happy_norm
```

```python
countryf=cd['country'].head()
h=happy_norm[0:5]
g=gdp_norm[0:5]
new=pd.DataFrame({'name':countryf,'gd':g,'hap':h})
new
```

```python
plt.figure(figsize=(7,4))
plt.scatter(gdp_norm,happy_norm)
plt.xlabel('GDP[normalised]')
plt.ylabel('Happiness Score [normalised]')
plt.legend(['Country'],loc='upper left')
plt.tight_layout()
for i,j in new.iterrows():
    plt.text(j['gd'],j['hap'],j['name'])
plt.show()
```
