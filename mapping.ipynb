```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Completed Items Combined.xlsx')
df.sample(6)
```

```python
df=df.set_index(['Core Ingredient Type','Sub-Type','Ingredient']).stack()
```

```python
df=df.to_frame()
```

```python
df=df.rename_axis(index=['Core Ingredient Type','Sub-Type','Ingredient','type'])
df=df.rename(columns={0:'Bool'})
df
```

```python
df=df.reset_index(level=[0,1,2,3])
```

```python
df.head()
```

```python
def isinlist2(y):
    return df[df['Ingredient']==y].type
```

```python
userask=df.Ingredient.to_list()
```

```python
var1=[]
ml=[]

x=input('Enter Ingredient: ')
print('')

liy=df[(df['Bool']=='Yes') & (df['Ingredient']==x)].type.values.tolist()
lin=df[(df['Bool']=='No') & (df['Ingredient']==x)].type.values.tolist()

if x in userask:
    var1.append(isinlist2(x).values)
    for i in var1:
        for j in i:
            ml.append(j)
else:
    print('Try again')
y=[]
n=[]
for i in ml:
    if i in liy:
        y.append(i)
for i in ml:
    if i in lin:
        n.append(i)
    
print('YES:',list(set(y)))
print('\n')
print('NO:',list(set(n)))
```
