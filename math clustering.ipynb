```python
# two different arrays
x = np.array([8,7,9,0])
y = np.array([-8,-7,15,16])
# needed for the graph below
my_range = np.arange(4)
# their respective means
mean_x = np.mean(x)
mean_y = np.mean(y)
fig, axs = plt.subplots(1,2, sharey=True)
axs[0].stem(my_range, x, label='x values', bottom=mean_x) 
axs[1].stem(my_range, y, label='y values', bottom=mean_y)
plt.show()
```

```python
std=my_data.std(0)
std
```

```python
plt.scatter(my_data[:,0],my_data[:,1])
plt.scatter(means[0],means[1])
```

```python
import matplotlib.patches as patch
ellipse=patch.Ellipse([means[0],means[1]],std[0]*2,std[1]*2,alpha=0.25)
fig,graphs=plt.subplots()
graphs.scatter(my_data[:,0],my_data[:,1])
graphs.scatter(means[0],means[1])
graphs.add_patch(ellipse)
```

```python
df=pd.read_csv('happyscore_income.csv')
df.info()
```

```python
df.sample(5)
```

```python
df.sort_values('avg_income',inplace=True)
```

```python
richest=df[df['avg_income']>15000]
richest.head()
```

```python
plt.scatter(richest['avg_income'],richest['happyScore'])
for k,i in richest.iterrows():
    plt.text(i['avg_income'],i['happyScore'],i['country'])
```

```python
plt.figure()
plt.xlabel('income')
plt.ylabel('happy score')
plt.scatter(df['avg_income'],df['happyScore'],s=df['income_inequality']*5,alpha=0.25)
plt.legend(['Inequality income country'],loc='upper left')
plt.show()
```

```python
happy=df['happyScore']
income=df['avg_income']
ineq=df['income_inequality']
```

```python
happy
```

```python
income
```

```python
from sklearn.cluster import KMeans


np.set_printoptions(suppress=True)
income_happy=np.column_stack((income,happy))

income_happy
```

```python
kmeanres=KMeans(n_clusters=3).fit(income_happy)
```

```python
kmeanres.cluster_centers_
```

```python
clustercen=kmeanres.cluster_centers_
plt.figure()
plt.scatter(income,happy)
plt.scatter(clustercen[:,0],clustercen[:,1],s=1200)
```
